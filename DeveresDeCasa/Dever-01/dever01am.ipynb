```python
import pandas as pd

# Função para ler o arquivo CSV
def ler_arquivo_csv(nome_arquivo):
    """Lê o arquivo CSV com separador ';' e codificação ISO-8859-1."""
    try:
        dados = pd.read_csv(nome_arquivo, sep=';', encoding='iso-8859-1')
        return dados
    except FileNotFoundError:
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return None

# Função para formatar o registro
def formatar_registro(registro):
    """Formata o registro conforme solicitado."""
    nome = registro["Nome"]
    data_nasc = pd.to_datetime(registro["Data de Nascimento"], format="%d/%m/%Y").strftime("%d/%m/%Y")
    data_cadastro = pd.to_datetime(registro["Dia do Cadastro"], format="%d/%m/%Y").strftime("%d/%m/%Y")
    hora_cadastro = registro["Hora do Cadastro"]
    return f"Nome: {nome}; Data de nascimento: {data_nasc}; Data de cadastro: {data_cadastro} às {hora_cadastro}"

# Função para ler um número inteiro válido
def ler_inteiro(strMensagem):
    """Lê um número inteiro válido da entrada do usuário."""
    while True:
        try:
            tamanho = int(input(strMensagem))
            return tamanho
        except ValueError:
            print("Erro: Por favor, digite um número inteiro válido.")

# Ler o arquivo CSV
nome_arquivo = "/content/drive/MyDrive/Colab Notebooks/dever_de_casa/cadastros.csv"  # Caminho no Colab
dados = ler_arquivo_csv(nome_arquivo)

if dados is not None:
    # Solicitar o índice do registro ao usuário
    indice = ler_inteiro("Digite o número do registro que deseja visualizar (0 a 4): ")

    # Verificar se o índice está dentro do intervalo
    if 0 <= indice < len(dados):
        registro = dados.iloc[indice]
        print(f"Registro {indice}: {formatar_registro(registro)}")
    else:
        print(f"Erro: Índice {indice} fora do intervalo. O arquivo contém apenas {len(dados)} registros.")
```
